```python
import gradio as gr
import random

#  Dummy prediction function (no model attached)
def scan_review(review_text):
    if not review_text.strip():
        return "⚠️ Please enter a review first.", 0.0

    # Dummy result — your team will replace this with real model later
    prob_fake = round(random.uniform(0.1, 0.95), 2)

    if prob_fake < 0.25:
        verdict = "✅ Non-AI (Human)"
        explanation = "Model strongly favors human-authored language patterns."
    elif prob_fake < 0.55:
        verdict = "⚠️ Uncertain"
        explanation = "Signals are mixed — manual review recommended."
    elif prob_fake < 0.80:
        verdict = "🚩 Likely AI-generated"
        explanation = "Linguistic markers align with automated text."
    else:
        verdict = "🚩 Surely AI-generated"
        explanation = "Model is highly confident this review is synthetic."

    pct = round(prob_fake * 100, 1)
    result = f"**{verdict}**\n\n{explanation}\n\n**AI Probability: {pct}%**"
    return result, pct


def scan_batch(batch_text):
    if not batch_text.strip():
        return "⚠️ Please enter at least one review.", None

    lines = [l.strip() for l in batch_text.strip().split("\n") if l.strip()]

    results = []
    fake_count = human_count = maybe_count = 0

    for i, line in enumerate(lines, 1):
        # Dummy result — replace with real model later
        prob_fake = round(random.uniform(0.1, 0.95), 2)
        pct = round(prob_fake * 100, 1)

        if prob_fake < 0.25:
            verdict = "✅ Human"
            human_count += 1
        elif prob_fake < 0.55:
            verdict = "⚠️ Uncertain"
            maybe_count += 1
        elif prob_fake < 0.80:
            verdict = "🚩 Likely AI"
            fake_count += 1
        else:
            verdict = "🚩 Surely AI"
            fake_count += 1

        results.append([i, line[:80] + ("…" if len(line) > 80 else ""), verdict, f"{pct}%"])

    total = len(results)
    summary = f"**Scanned {total} reviews** — 🚩 AI/Fake: {fake_count} | ⚠️ Uncertain: {maybe_count} | ✅ Human: {human_count}"
    return summary, results


# Gradio UI
with gr.Blocks(
    title="Fake Review Detector",
    theme=gr.themes.Base(
        primary_hue="red",
        neutral_hue="slate",
        font=gr.themes.GoogleFont("Syne"),
    ),
    css="""
    #title { font-size: 30px; font-weight: 800; margin-bottom: 6px; }
    #subtitle { color: #888; font-size: 14px; margin-bottom: 24px; }
    """
) as demo:

    gr.HTML("<h1 id='title'>🔍 Fake Review Detector</h1>")
    gr.HTML("<p id='subtitle'>Detects AI-generated fake reviews — Frontend UI (model integration coming soon)</p>")

    with gr.Tabs():

        # Tab 1: Single Review
        with gr.TabItem("Single Review"):
            gr.Markdown("Paste any product or restaurant review below and hit **Scan**.")
            with gr.Row():
                with gr.Column():
                    single_input = gr.Textbox(
                        label="Review Text",
                        placeholder="e.g. This product is amazing! Fast shipping and great quality...",
                        lines=6
                    )
                    scan_btn = gr.Button("Scan Review", variant="primary", size="lg")

                with gr.Column():
                    single_output = gr.Markdown(label="Result")
                    prob_slider = gr.Slider(
                        label="AI Probability %",
                        minimum=0, maximum=100, value=0,
                        interactive=False
                    )

            scan_btn.click(
                fn=scan_review,
                inputs=single_input,
                outputs=[single_output, prob_slider]
            )

        # Tab 2: Batch Scan
        with gr.TabItem("Batch Scan"):
            gr.Markdown("Paste multiple reviews below, **one per line**, then hit Scan All.")
            batch_input = gr.Textbox(
                label="Reviews (one per line)",
                placeholder="Review 1...\nReview 2...\nReview 3...",
                lines=10
            )
            batch_btn = gr.Button("Scan All", variant="primary", size="lg")
            batch_summary = gr.Markdown()
            batch_table = gr.Dataframe(
                headers=["#", "Review", "Verdict", "AI Probability"],
                label="Results",
                wrap=True
            )

            batch_btn.click(
                fn=scan_batch,
                inputs=batch_input,
                outputs=[batch_summary, batch_table]
            )

        # Tab 3: About
        with gr.TabItem("About"):
            gr.Markdown("""
""")
```
